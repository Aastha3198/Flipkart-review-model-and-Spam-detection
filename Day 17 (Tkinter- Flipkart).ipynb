```python
import tkinter as tk
import joblib
import pickle
import pandas as pd
from nltk.stem.porter import PorterStemmer

root=tk.Tk()
root.title("Flipkart review model")
root.geometry('700x650')
root.config(bg='#d9d9d9')

def review_prediction():
    rating=rate_input.get(0.0,tk.END)
    review=review_input.get(0.0,tk.END)
    summary=summary_input.get(0.0,tk.END)

    rating = list(rating)
    review = [''.join(review.split())]
    summary = summary.split()

    tag = rating + review + summary

    porter_stemer = PorterStemmer()

    temp=[]
    for word in tag:
        result=porter_stemer.stem(word)
        temp.append(result)

    tag=' '.join(temp)
    temp.clear()

    tag = pd.Series(tag)
    vector_model = joblib.load('Vector Model.joblib')
    review_model = pickle.load(open('Flipkart Review Model.sav', 'rb'))
    tag = vector_model.transform(tag)
    prediction = review_model.predict(tag)
    output.config(text=prediction[0].capitalize())

heading=tk.Label(text='Product Review',font=('Kristen ITC',50),fg='#ef233c',bg='#d9d9d9')
heading.grid(row=1,column=1,columnspan=2,padx=80,pady=20)

tk.Label(text='Rate: ',font=('Kristen ITC',25),bg='#d9d9d9').grid(row=2,column=1)
tk.Label(text='Review: ',font=('Kristen ITC',25),bg='#d9d9d9').grid(row=3,column=1)
tk.Label(text='Summary: ',font=('Kristen ITC',25),bg='#d9d9d9').grid(row=4,column=1,columnspan=2)

# Use text instead of entry to give border to the box
rate_input=tk.Text(font=('Arial',20),width=20,height=1,relief='solid')
review_input=tk.Text(font=('Arial',20),width=20,height=1,relief='solid')
summary_input=tk.Text(font=('Arial',15),width=55,height=6,relief='solid')
rate_input.grid(row=2,column=2)
review_input.grid(row=3,column=2)
summary_input.grid(row=5,column=1,columnspan=2)

btn=tk.Button(text='Predict',font=('Kristen ITC',20,'bold'),bg='#ef233c',fg='white',width=32,relief='solid',command=review_prediction)
btn.grid(row=6,column=1,columnspan=2,pady=15)

output=tk.Label(text='',font=('Kristen ITC',50),fg='#ef233c',bg='#d9d9d9')
output.grid(row=7,column=1,columnspan=2,padx=80,pady=10)

root.mainloop()
```
